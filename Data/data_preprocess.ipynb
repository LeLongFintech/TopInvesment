# DATA PROCESSING

```python
id_cols = ["Tickers", "Date", "IPO Date", "GICS Industry Name"]
num_cols = [c for c in balance_sheet.columns if c not in id_cols]

bs = balance_sheet.copy()
```

```python
bs["Date"] = pd.to_datetime(bs["Date"], errors="coerce")
bs["IPO Date"] = pd.to_datetime(bs["IPO Date"], errors="coerce")
bs["Tickers"] = bs["Tickers"].astype(str).str.strip()
bs["GICS Industry Name"] = bs["GICS Industry Name"].astype(str).str.strip()

for col in num_cols:
    bs[col] = pd.to_numeric(bs[col], errors="coerce")

bs.dtypes
```

```python
dup_count = bs.duplicated(subset=["Tickers", "Date"]).sum()
print(f"Duplicates (Tickers + Date): {dup_count}")

bs = bs.drop_duplicates(subset=["Tickers", "Date"], keep="last")
bs.shape
```

```python
bs = bs.sort_values(["Tickers", "Date"]).reset_index(drop=True)
```

```python
missing = bs[num_cols].isnull().sum()
missing_pct = (missing / len(bs) * 100).round(2)
missing_df = pd.DataFrame({"count": missing, "%": missing_pct})
missing_df = missing_df[missing_df["count"] > 0].sort_values("%", ascending=False)
display(missing_df)
print(f"\nTotal rows: {len(bs)}")
```

```python
threshold = 50
high_missing = missing_pct[missing_pct > threshold].index.tolist()
print(f"Columns dropped (>{threshold}% missing): {high_missing}")

bs = bs.drop(columns=high_missing)
num_cols = [c for c in num_cols if c not in high_missing]
```

```python
ticker_counts = bs.groupby("Tickers")["Date"].count()
min_periods = 4  # ít nhất 4 quý (1 năm)

low_data_tickers = ticker_counts[ticker_counts < min_periods].index.tolist()
print(f"Tickers removed (< {min_periods} periods): {len(low_data_tickers)}")

bs = bs[~bs["Tickers"].isin(low_data_tickers)].reset_index(drop=True)
bs.shape
```

```python
bs[num_cols] = bs.groupby("Tickers")[num_cols].transform(
    lambda x: x.ffill().bfill()
)

remaining_missing = bs[num_cols].isnull().sum()
remaining_missing = remaining_missing[remaining_missing > 0]
print(f"Remaining missing columns: {len(remaining_missing)}")
if len(remaining_missing) > 0:
    display(remaining_missing)
```

```python
still_missing = bs[num_cols].isnull().any()
cols_still_missing = still_missing[still_missing].index.tolist()

if cols_still_missing:
    bs[cols_still_missing] = bs.groupby("GICS Industry Name")[cols_still_missing].transform(
        lambda x: x.fillna(x.median())
    )

# Nếu vẫn còn missing (industry cũng toàn NaN) -> fill median toàn bộ
remaining = bs[num_cols].isnull().sum()
remaining = remaining[remaining > 0]
if len(remaining) > 0:
    bs[num_cols] = bs[num_cols].fillna(bs[num_cols].median())

print(f"Total missing after all fills: {bs[num_cols].isnull().sum().sum()}")
```

```python
pd.DataFrame({
    "dtype": bs[num_cols].dtypes,
    "missing": bs[num_cols].isnull().sum(),
    "zeros": (bs[num_cols] == 0).sum(),
    "negatives": (bs[num_cols] < 0).sum(),
    "mean": bs[num_cols].mean().round(2),
    "std": bs[num_cols].std().round(2),
    "min": bs[num_cols].min().round(2),
    "max": bs[num_cols].max().round(2)
})
```

```python
print(f"Before: {balance_sheet.shape}")
print(f"After:  {bs.shape}")
print(f"Rows removed: {balance_sheet.shape[0] - bs.shape[0]}")
print(f"Cols removed: {balance_sheet.shape[1] - bs.shape[1]}")
print(f"Missing remaining: {bs.isnull().sum().sum()}")
```

```python
eq = eps_quarter.copy()

eq.columns = eq.columns.str.strip()
eq["Tickers"] = eq["Tickers"].astype(str).str.strip()
eq["Date"] = pd.to_datetime(eq["Date"], errors="coerce")
eq["Common Shares - Outstanding - Total"] = pd.to_numeric(eq["Common Shares - Outstanding - Total"], errors="coerce")
eq["Net Income after Tax"] = pd.to_numeric(eq["Net Income after Tax"], errors="coerce")

eq.dtypes
```

```python
print("Missing:")
print(eq.isnull().sum())
print(f"\nShares <= 0: {(eq['Common Shares - Outstanding - Total'] <= 0).sum()}")
print(f"Shares NaN:  {eq['Common Shares - Outstanding - Total'].isna().sum()}")
print(f"Net Income NaN: {eq['Net Income after Tax'].isna().sum()}")
```

```python
# Shares phải > 0 để tính EPS, không thể fill
before = len(eq)
eq = eq.dropna(subset=["Common Shares - Outstanding - Total"])
eq = eq[eq["Common Shares - Outstanding - Total"] > 0]
print(f"Rows removed (invalid shares): {before - len(eq)}")

# Net Income: fill forward theo ticker (quý trước gần nhất)
eq["Net Income after Tax"] = eq.groupby("Tickers")["Net Income after Tax"].transform(
    lambda x: x.ffill().bfill()
)

# Drop nếu vẫn còn NaN
remaining_na = eq.isnull().sum().sum()
if remaining_na > 0:
    eq = eq.dropna().reset_index(drop=True)

print(f"Final rows: {len(eq)}, Missing: {eq.isnull().sum().sum()}")
```

```python
raw = price.copy()
print(f"Shape: {raw.shape}")
print("Header (columns):", raw.columns[:6].tolist())
print("Row 0:", raw.iloc[0, :6].tolist())
print("Row 1:", raw.iloc[1, :6].tolist())
print("Row 2:", raw.iloc[2, :6].tolist())
```

```python
# Bỏ dòng 0 (chứa tên biến), data bắt đầu từ dòng 1
data = raw.iloc[1:].copy().reset_index(drop=True)

# Cột đầu = Date
data["Unnamed: 0"] = pd.to_datetime(data["Unnamed: 0"], errors="coerce")
data = data.rename(columns={"Unnamed: 0": "Date"})

# Tách: cột không có .1 -> Close, cột có .1 -> Volume
all_cols = data.columns[1:]
close_cols = [c for c in all_cols if not c.endswith(".1")]
volume_cols = [c for c in all_cols if c.endswith(".1")]

df_close = data[["Date"] + close_cols].copy()
df_volume = data[["Date"] + volume_cols].copy()

# Bỏ .1 ở volume để tên ticker khớp
df_volume.columns = ["Date"] + [c[:-2] for c in volume_cols]

# Convert numeric
for col in df_close.columns[1:]:
    df_close[col] = pd.to_numeric(df_close[col], errors="coerce")
for col in df_volume.columns[1:]:
    df_volume[col] = pd.to_numeric(df_volume[col], errors="coerce")

print(f"df_close:  {df_close.shape}")
print(f"df_volume: {df_volume.shape}")
display(df_close.iloc[:3, :5])
display(df_volume.iloc[:3, :5])
```

```python
df_close = df_close.sort_values("Date").reset_index(drop=True)
df_volume = df_volume.sort_values("Date").reset_index(drop=True)
```

```python
close_tickers = set(df_close.columns[1:])
volume_tickers = set(df_volume.columns[1:])

keep = list(close_tickers & volume_tickers & common_tickers)

df_close = df_close[["Date"] + keep]
df_volume = df_volume[["Date"] + keep]

print(f"Tickers kept: {len(keep)}")
print(f"df_close:  {df_close.shape}")
print(f"df_volume: {df_volume.shape}")
```

```python
print(f"Missing close:  {df_close.iloc[:, 1:].isnull().sum().sum()}")
print(f"Missing volume: {df_volume.iloc[:, 1:].isnull().sum().sum()}")

# Price: ffill theo thời gian (giữ giá gần nhất nếu không giao dịch)
df_close.iloc[:, 1:] = df_close.iloc[:, 1:].ffill()
# Volume: không giao dịch = 0
df_volume.iloc[:, 1:] = df_volume.iloc[:, 1:].fillna(0)

print(f"\nAfter fill:")
print(f"Missing close:  {df_close.iloc[:, 1:].isnull().sum().sum()}")
print(f"Missing volume: {df_volume.iloc[:, 1:].isnull().sum().sum()}")
```

```python
# Bfill phần đầu chuỗi nếu còn NaN
df_close.iloc[:, 1:] = df_close.iloc[:, 1:].bfill()

remaining = df_close.isnull().sum().sum()
print(f"Final missing close: {remaining}")
print(f"Final missing volume: {df_volume.isnull().sum().sum()}")
```

```python
print(f"df_close:  {df_close.shape} | {df_close['Date'].min()} -> {df_close['Date'].max()}")
print(f"df_volume: {df_volume.shape} | {df_volume['Date'].min()} -> {df_volume['Date'].max()}")
display(df_close.head())
```

```python
cutoff = pd.Timestamp("2025-12-31")

df_close = df_close[df_close["Date"] <= cutoff].reset_index(drop=True)
df_volume = df_volume[df_volume["Date"] <= cutoff].reset_index(drop=True)

print(f"df_close:  {df_close.shape} | {df_close['Date'].min()} -> {df_close['Date'].max()}")
print(f"df_volume: {df_volume.shape} | {df_volume['Date'].min()} -> {df_volume['Date'].max()}")
```

```python
common_tickers = set(keep)  # cập nhật lại common_tickers

bs = bs[bs["Tickers"].isin(common_tickers)].reset_index(drop=True)
ist = ist[ist["Tickers"].isin(common_tickers)].reset_index(drop=True)
cf = cf[cf["Tickers"].isin(common_tickers)].reset_index(drop=True)
eq = eq[eq["Tickers"].isin(common_tickers)].reset_index(drop=True)

print(f"balance_sheet:    {bs.shape}  |  {bs['Tickers'].nunique()} tickers")
print(f"income_statement: {ist.shape}  |  {ist['Tickers'].nunique()} tickers")
print(f"cash_flow:        {cf.shape}  |  {cf['Tickers'].nunique()} tickers")
print(f"eps_quarter:      {eq.shape}  |  {eq['Tickers'].nunique()} tickers")
print(f"df_close:         {df_close.shape}  |  {df_close.shape[1]-1} tickers")
print(f"df_volume:        {df_volume.shape}  |  {df_volume.shape[1]-1} tickers")
```
