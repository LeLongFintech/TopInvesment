```python
import pandas as pd
import numpy as np
```

```python
bs_2000_2015 = pd.read_excel(r".\Data\raw data\Balance Sheet\Balance_sheet(2000 - 2015).xlsx")
bs_2015_2025 = pd.read_excel(r".\Data\raw data\Balance Sheet\Balance_sheet(2015 - 2025).xlsx")
info = pd.read_excel(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 7\Phân tích chứng khoán\Project\Data\Thailand_market_info.xlsx")
```

```python
cf_2000_2015 = pd.read_excel(r"./Data/raw data/Cash Flow/Cash_flow(2000 - 2015).xlsx")
cf_2015_2025 = pd.read_excel(r"./Data/raw data/Cash Flow/Cash_flow(2015 - 2025).xlsx")
```

```python
is_2000_2015 = pd.read_excel(r"./Data/raw data/Income Statement/Income_statement(2000 - 2015).xlsx")
is_2015_2025 = pd.read_excel(r"./Data/raw data/Income Statement/Income_statement(2015 - 2025).xlsx")
```

```python
info
```

```python
def preprocess_balance_sheet(df):
    df = df.copy()
    df.columns = df.columns.str.strip()
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Tickers', 'Date'])
    df = df.drop_duplicates(subset=['Tickers', 'Date'], keep='last')
    return df

bs_2000_2015 = preprocess_balance_sheet(bs_2000_2015)
bs_2015_2025 = preprocess_balance_sheet(bs_2015_2025)
cf_2000_2015 = preprocess_balance_sheet(cf_2000_2015)
cf_2015_2025 = preprocess_balance_sheet(cf_2015_2025)
is_2000_2015 = preprocess_balance_sheet(is_2000_2015)
is_2015_2025 = preprocess_balance_sheet(is_2015_2025)
```

```python
bs_merged = (
    pd.concat([bs_2000_2015, bs_2015_2025], ignore_index=True)
    .drop_duplicates(subset=['Tickers', 'Date'], keep='last')
    .sort_values(by=['Tickers', 'Date'])
    .reset_index(drop=True)
)
cf_merged = (
    pd.concat([cf_2000_2015, cf_2015_2025], ignore_index=True)
    .drop_duplicates(subset=['Tickers', 'Date'], keep='last')
    .sort_values(by=['Tickers', 'Date'])
    .reset_index(drop=True)
)
is_merged = (
    pd.concat([is_2000_2015, is_2015_2025], ignore_index=True)
    .drop_duplicates(subset=['Tickers', 'Date'], keep='last')
    .sort_values(by=['Tickers', 'Date'])
    .reset_index(drop=True)
)
```

```python
info.columns = info.columns.str.strip()

info_cols = info[['Tickers', 'IPO Date', 'GICS Industry Name']]

bs_merged = bs_merged.merge(info_cols, on='Tickers', how='left')
cf_merged = cf_merged.merge(info_cols, on='Tickers', how='left')
is_merged = is_merged.merge(info_cols, on='Tickers', how='left')
```

```python
bs_merged.to_excel(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 7\Phân tích chứng khoán\Project\Data\raw data\Balance Sheet\Balance_sheet.xlsx", index=False)
cf_merged.to_excel(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 7\Phân tích chứng khoán\Project\Data\raw data\Cash Flow\Cash_flow.xlsx", index=False)
is_merged.to_excel(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 7\Phân tích chứng khoán\Project\Data\raw data\Income Statement\Income_statement.xlsx", index=False)
```
